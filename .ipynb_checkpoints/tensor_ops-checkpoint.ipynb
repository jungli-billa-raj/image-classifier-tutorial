```python
import torch 
```

```python
my_torch = torch.arange(10)
my_torch
```

```python
# reshape
my_torch = my_torch.reshape(2, 5)
# my_torch = my_torch.reshape(2, 6) # error 
my_torch
```

```python
# reshape without knowing the size using -1. 
my_torch = torch.arange(10)
my_torch = my_torch.reshape(5, -1)
my_torch
```
