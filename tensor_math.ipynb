```python
import torch 
import numpy as np
```

```python
tensor_a = torch.tensor([0,1,2,3,4])
tensor_b = torch.tensor([5,6,7,8,9])
print(tensor_a)
print(tensor_b)
```

```python
# ADDITION Shorthand form 
tensor_a + tensor_b 
```

```python
# ADDITION Long form 
# I feel like this approach is definitely more expressive and clearly shows that tensors are being worked upon. 
torch.add(tensor_a , tensor_b)
```

```python
# SUBTRACTION Shorthand form 
tensor_b - tensor_a 
```

```python
# SUBTRACTION Long form 
torch.sub(tensor_b, tensor_a) # could also use .subtract()
```

```python
# MULTIPLICATION short form 
tensor_a * tensor_b
```

```python
# MULTIPLICATION long form 
torch.mul(tensor_a, tensor_b)
```

```python
# DIVISION short form 
tensor_b / tensor_a 
```

```python
# DIVISION long form 
torch.div(tensor_b, tensor_a)
```

```python
# REMAINDER short form 
# tensor_b % tensor_a # ZeroDivisionError
tensor_a % tensor_b 
```

```python
# REMAINDER long form 
torch.remainder(tensor_a, tensor_b) 
```

```python
# EXPONENT short form 
tensor_b ** tensor_a
```

```python
# EXPONENT long form 
torch.pow(tensor_b, tensor_a) 
```
